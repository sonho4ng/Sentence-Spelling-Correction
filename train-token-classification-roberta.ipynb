```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/kaggle/input/luat123/error_dataset.csv")
data
```

```python
def split_token(text):
    return text.strip().split()
def split_label(mask_string):
    return [int(char) for char in mask_string.strip()]

data['corrupted'] = data['corrupted'].apply(split_token)
data['mask'] = data['mask'].apply(split_label)
data
```

```python
# Dataset return {token_ids, attention_masks, labels} c·ªßa t·ª´ng token trong sentence ƒë√≥
# define tokenizer, PhoBERT AutoModelForTokenClassification th√¨ ko c·∫ßn g·ªçi loss
# g·ªçi dataset, chia b·∫±ng dataloader
# optim l√† Adam
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch.nn.functional as F
import torch.optim as optim
from sklearn.metrics import f1_score
```

```python
class ErrorDetectionDataset(Dataset):
    def __init__(self, data, tokenizer, max_len=128):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        tokens = self.data.iloc[idx]["corrupted"]
        labels = self.data.iloc[idx]["mask"]

        encoding = self.tokenizer(tokens,
                                  is_split_into_words=True,
                                  padding="max_length",
                                  truncation=True,
                                  max_length=self.max_len,
                                  return_tensors="pt")
        
        word_ids = encoding.word_ids()
        label_ids = []
        prev_word_idx = None

        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != prev_word_idx:
                label_ids.append(labels[word_idx])
            else:
                label_ids.append(labels[word_idx])  # ho·∫∑c -100 n·∫øu kh√¥ng label subtokens
            prev_word_idx = word_idx

        encoding = {k: v.squeeze(0) for k, v in encoding.items()}
        encoding["labels"] = torch.tensor(label_ids)

        return encoding
```

```python
model_name = "xlm-roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
model = AutoModelForTokenClassification.from_pretrained(model_name, num_labels=2)
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(data, test_size=1000, random_state=42)
train_dataset = ErrorDetectionDataset(train_df, tokenizer)
val_dataset = ErrorDetectionDataset(val_df, tokenizer)
```

```python
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
```

```python
from tqdm import tqdm
import os
import torch
import torch.optim as optim

from sklearn.metrics import classification_report, f1_score, precision_score, recall_score
best_eval_loss = float("inf")  # Kh·ªüi t·∫°o loss t·ªët nh·∫•t ban ƒë·∫ßu

save_path = "checkpoint"
os.makedirs(save_path, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
optimizer = optim.Adam(model.parameters(), lr=2e-5)
epochs = 4

for epoch in range(epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    # ‚úÖ tqdm here
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}", leave=False)

    for batch in progress_bar:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        total_loss += loss.item()

        # Flatten predictions and labels for F1
        preds = torch.argmax(logits, dim=-1)
        mask = labels != -100
        preds = preds[mask].detach().cpu().numpy()
        true = labels[mask].detach().cpu().numpy()

        all_preds.extend(preds)
        all_labels.extend(true)

        progress_bar.set_postfix(loss=loss.item())

    avg_loss = total_loss / len(train_loader)
    precision = precision_score(all_labels, all_preds, pos_label=1)
    recall = recall_score(all_labels, all_preds, pos_label=1)
    f1 = f1_score(all_labels, all_preds, pos_label=1)
    precision2 = precision_score(all_labels, all_preds, pos_label=0)
    recall2 = recall_score(all_labels, all_preds, pos_label=0)
    f12 = f1_score(all_labels, all_preds, pos_label=0)
    print(f"\nEpoch {epoch+1} - Loss: {avg_loss:.4f} - F1: {f1:.4f} - Recall: {recall:.4f} - Precision: {precision:.4f} - F12: {f12:.4f} - Recall2: {recall2:.4f} - Precision2: {precision2:.4f}")
    # th√™m eval
    eval_loss = evaluate(model, val_loader)

    # ‚úÖ Save model n·∫øu eval_loss t·ªët nh·∫•t
    if eval_loss < best_eval_loss:
        best_eval_loss = eval_loss
        model.save_pretrained(f"{save_path}/best")
        tokenizer.save_pretrained(f"{save_path}/best")
        print(f"‚úÖ Model improved! Saved to {save_path}/best")
    else:
        print("No improvement, model not saved.")
    

```
